```python
class Product():
    def __init__(self,name,price):
       self.name = name
       self.price = price
    
    def display_info(self):
        print(f"{self.name}----{self.price}")

class ShoppingCart():
    def __init__(self):
        self.items={}
        
    def add_product(self, product, quantity=1):
        if product.name in self.items:
            self.items[product.name]["quantity"] += quantity
        else:
            self.items[product.name] = {"product": product, "quantity": quantity}
            
    def remove_product(self,product,quantity=1):
        if product.name in self.items:
            self.items[product.name]["quantity"] -= quantity
            if self.items[product.name]["quantity"] <=0:
                del self.items[product.name]
            
        else:
            print(f"you have no {product.name} registered in your cart")

    def total_cost(self):
        total=0
        for value in self.items.values():
            total += value["product"].price * value["quantity"]
        print(f"Your total is {total:.2f} euros")
        return total
           

    def display_cart(self):
        print("\nShopping Cart:")
        for name, data in self.items.items():
            print(f"{name}: {data['quantity']} units - ${data['product'].price:.2f} each")
        self.total_cost()
           
    
```

```python

# Test the implementation
# Create products
apple = Product("Apple", 0.99)
banana = Product("Banana", 0.59)
milk = Product("Milk", 3.49)
apple.display_info()
banana.display_info()
milk.display_info()
# Create cart
cart = ShoppingCart()

# Add items
cart.add_product(apple, 3)
cart.add_product(banana)
cart.add_product(milk, 2)
cart.display_cart()

# Remove items
cart.remove_product(apple, 1)
cart.remove_product(banana)
cart.display_cart()
cart.total_cost()

# Try to remove a product not in the cart
# cart.remove_product(orange,1)
cart.remove_product(milk, 5)  # Removes all milk
cart.display_cart()
```
